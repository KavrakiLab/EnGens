```python
import pandas as pd
import os
import ast
```

```python
infolder = "."
file_prefix = "cluster_centers_results"
file_ext = ".csv"
dfs = []
for file in os.listdir(infolder):
    if file[-len(file_ext):]==file_ext and file_prefix in file:
        print(file)
        file_label = file[len(file_prefix)+1:-len(file_ext)]
        print(file_label)
        tmp_df = pd.read_csv(file)
        tmp_df["label"] = file_label
        dfs.append(tmp_df)
        
input_representatives = pd.concat(dfs)
```

```python
input_representatives["frame_number_list"] = input_representatives["frame_number"].apply(lambda x: ast.literal_eval(x))
```

```python
experiment = ("pca-kmeans", "pairwise distance to 4 points")

indata = input_representatives[input_representatives["label"]==experiment[0]]
indata = indata[indata["f_name"]==experiment[1]]
indata
```
