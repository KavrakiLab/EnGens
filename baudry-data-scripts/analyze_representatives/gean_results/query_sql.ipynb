```python
# get number of actives in top X
def get_actives_num(frame_num, mydb):
    proc = {"05":None, 
            "1":None, 
            "5":None, 
            "10":None}

    for p in proc.keys():
        print("running query")
        mycursor = mydb.cursor()
        mycursor.execute("SELECT COUNT(*) FROM unique_actives_per_frame{} WHERE frame_num={}".format(p, frame_num))
        myresult = mycursor.fetchone()
        proc[p] = myresult
    return proc
```

## How many missing frame data do we have?

```python
## Check how many frames have missing data from the score list
# get number of actives in top X
def get_count_top05(frame_num, mydb):
    mycursor = mydb.cursor()
    mycursor.execute("SELECT COUNT(*) FROM adora2a_top05 WHERE frame_num={}".format(frame_num))
    myresult = mycursor.fetchone()
    return myresult
```

```python
frame_map
```

```python
missing_frames = []
missing_pdb_ind = []
for row in frame_map.iterrows():
    print("Frame: ")
    print(row[1]["file_index"])
    pdb_ind = row[1]["file_index"]
    cnt = get_count_top05(pdb_ind, mydb)
    print("Cnt top05:")
    print(cnt)
    if cnt[0] == 0:
        missing_frames.append(row[1]["traj_ind"])
        missing_pdb_ind.append(pdb_ind)
```

```python
missing_frames = pd.DataFrame({"frame_ind": missing_frames, "pdb_ind": missing_pdb_ind})
```

```python
missing_frames.to_csv("missing_frames.csv", index=False)
```

## For missing data - take the closest frame that we have data for

```python
ens_frames
```

```python
res_df = {"orig_frame": [], "closest_frame": []}
for row in ens_frames.iterrows():
    r = row[1]
    if r["missing_dock_data"] == 1:
        found_closest = False
        # get closest frame
        step = 0
        while not found_closest:
            traj_ind = r["frame_id"]
            step+=1
            #try one up one down
            new_ti = traj_ind + step
            if new_ti in list(missing_frames["frame_ind"]):
                new_ti = traj_ind - step
                if not new_ti in list(missing_frames["frame_ind"]):
                    break
            else:
                break
        closest_frame = new_ti
        res_df["orig_frame"].append(r["frame_id"])
        res_df["closest_frame"].append(closest_frame)
```

```python
res_pd = pd.DataFrame(res_df)
res_pd
```

```python
res_pd.to_csv("missing_closest_found.csv", index=False)
```
